```python
import os

# Caminho da pasta raiz da solução
solution_directory = "E:/GitDesktop/SommusGestor-web/src"

# Configuração da substituição
old_framework = "<TargetFrameworkVersion>v4.6.1</TargetFrameworkVersion>"
new_framework = "<TargetFrameworkVersion>v4.6.2</TargetFrameworkVersion>"

# Função para atualizar o conteúdo dos arquivos .csproj
def update_csproj_files(directory):
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".csproj"):
                file_path = os.path.join(root, file)

                # Ler o conteúdo do arquivo
                with open(file_path, "r", encoding="utf-8") as f:
                    content = f.read()

                # Substituir a versão do framework
                if old_framework in content:
                    updated_content = content.replace(old_framework, new_framework)

                    # Salvar as alterações no arquivo
                    with open(file_path, "w", encoding="utf-8") as f:
                        f.write(updated_content)

                    print(f"Atualizado: {file_path}")
                else:
                    print(f"Sem alterações: {file_path}")

# Executar a atualização
update_csproj_files(solution_directory)
```
